```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
#importing dataset
dataset=pd.read_csv('User_Data.csv')
X=dataset.iloc[:,2:4].values
Y=dataset.iloc[:,4].values
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.fit_transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
#Making confusion Matrix
cm=confusion_matrix(Y_pred, Y_test)
print(cm)
```

## Decision Tree

```python
#load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#working with dataset
col_names=["company","job","degree","salary_more_than_100K"]
data=pd.read_csv("salaries.csv",header=None,names=col_names)
```

```python
data
```

```python
#split data into features and target variables
feature_col=["company","job","degree"]
X=data[feature_col]
Y=data['salary_more_than_100K']
```

```python
#import labbel encoders
from sklearn import preprocessing
#label_encoder object knows how to understand word labels
label_encoder = preprocessing.LabelEncoder()
#Encode labels in columns
data["company"]=label_encoder.fit_transform(data["company"])
data["job"]=label_encoder.fit_transform(data["job"])
data["degree"]=label_encoder.fit_transform(data["degree"])
print(data.head())
```

```python
#split data into features and target variables
feature_col=["company","job","degree"]
X=data[feature_col] 
Y=data['salary_more_than_100K']  
#X=data.values[1:,:3] 
#Y=data.values[1:,3] #1:,3 means we are not using header
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size = 0.2, random_state=100)
```

```python
#Create a Decision Tree classifier object using entropy
clf_entropy=DecisionTreeClassifier(criterion="entropy",max_depth=2)
```

```python
#train Decision Tree Classifier
clf_entropy=clf_entropy.fit(X_train,Y_train)
```

```python
#predict the response for dataset
Y_pred=clf_entropy.predict(X_test)
print("Accuracy", metrics.accuracy_score(Y_test,Y_pred))
```
